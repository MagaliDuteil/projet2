```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

## Création des DataFrames

## Selectionner l'une des 3 villes (New york city, Washington ou Chicago)
    => Choix du df correspondasnt à la ville souhaitée

```python
df_nyc = pd.read_csv("C:/Users/Magal/OneDrive/Bureau/cleaned_new_york_city.csv")
df_nyc = pd.DataFrame(df_nyc) # New York City

df_wash = pd.read_csv("C:/Users/Magal/OneDrive/Bureau/cleaned_washington_city.csv")
df_wash = pd.DataFrame(df_wash) # Washington

df_chi = pd.read_csv("C:/Users/Magal/OneDrive/Bureau/cleaned_chicago_city.csv")
df_chi = pd.DataFrame(df_chi) # Chicago
```

# Création d'une colone mois par ville

```python
# Extraction du mois à partir du datetime
# Convertir la colonne "Start Time" en datetime si elle ne l'est pas déjà
df_nyc['Start Time'] = pd.to_datetime(df_nyc['Start Time'])
df_nyc['End Time'] = pd.to_datetime(df_nyc['End Time'])
# Extraction du mois à partir du datetime
df_nyc['Month'] = df_nyc['Start Time'].dt.month


df_wash['Start Time'] = pd.to_datetime(df_wash['Start Time'])
df_wash['End Time'] = pd.to_datetime(df_wash['End Time'])
df_wash['Month'] = df_wash['Start Time'].dt.month


df_chi['Start Time'] = pd.to_datetime(df_chi['Start Time'])
df_chi['End Time'] = pd.to_datetime(df_chi['End Time'])
df_chi['Month'] = df_chi['Start Time'].dt.month
```

# Le jour de la semaine avec le plus d'activité.

Création d'une colone jour par ville

```python
# Extraction du jour à partir du datetime
df_nyc['Day_of_Week'] = df_nyc['Start Time'].dt.strftime('%A')

df_wash['Day_of_Week'] = df_wash['Start Time'].dt.strftime('%A')

df_chi['Day_of_Week'] = df_chi['Start Time'].dt.strftime('%A')
```

Création d'un tableau de fréquentation par jour de la semaine

```python
df_nyc_fday = df_nyc.groupby('Day_of_Week').size().sort_values(ascending = False)
print(df_nyc_fday)
```

```python
df_wash_fday = df_wash.groupby('Day_of_Week').size().sort_values(ascending = False)
print(df_wash_fday)
```

```python
df_chi_fday = df_chi.groupby('Day_of_Week').size().sort_values(ascending = False)
print(df_chi_fday)
```

## L'heure de démarrage la plus courante.

création d'une colone heure de démarrage par ville

```python
# Extraction de l'heurer à partir du datetime
df_nyc['Hour'] = df_nyc['Start Time'].dt.hour

df_wash['Hour'] = df_wash['Start Time'].dt.hour

df_chi['Hour'] = df_chi['Start Time'].dt.hour
```

création d'un tableau de fréquentation par heure de la journée (heure de démarrage)

```python
df_nyc_fhour = df_nyc.groupby('Hour').size().sort_values(ascending = False)
print(df_nyc_fhour)
```

```python
df_wash_fhour = df_wash.groupby('Hour').size().sort_values(ascending = False)
print(df_wash_fhour)
```

```python
df_chi_fhour = df_chi.groupby('Hour').size().sort_values(ascending = False)
print(df_chi_fhour)
```

## La durée de voyage moyen sur la période (mois).

```python
df_nyc_meanm = df_nyc.groupby('Month')['Trip Duration'].mean() # Calcul de la durée moyenne de la location par mois
df_nyc_meanm= df_nyc_meanm.sort_values(ascending=False) # Tri par durée la plus élevée
df_nyc_meanm = df_nyc_meanm.apply(lambda x: divmod(x, 3600)) # Convertir la moyenne en heures et en secondes
df_nyc_meanm = df_nyc_meanm.apply(lambda x: divmod(x[1], 60)) 
df_nyc_meanm = df_nyc_meanm.apply(lambda x: '{:02}:{:02}:00'.format(int(x[0]), int(x[1]))) # Formatetr en heure : minutes : secondes

print(df_nyc_meanm)
```

```python
df_wash_meanm = df_wash.groupby('Month')['Trip Duration'].mean() 
df_wash_meanm= df_wash_meanm.sort_values(ascending=False) 
df_wash_meanm = df_wash_meanm.apply(lambda x: divmod(x, 3600)) 
df_wash_meanm = df_wash_meanm.apply(lambda x: divmod(x[1], 60)) 
df_wash_meanm = df_wash_meanm.apply(lambda x: '{:02}:{:02}:00'.format(int(x[0]), int(x[1])))
print(df_wash_meanm)
```

```python
df_chi_meanm = df_chi.groupby('Month')['Trip Duration'].mean() 
df_chi_meanm= df_chi_meanm.sort_values(ascending=False) 
df_chi_meanm = df_chi_meanm.apply(lambda x: divmod(x, 3600)) 
df_chi_meanm = df_chi_meanm.apply(lambda x: divmod(x[1], 60)) 
df_chi_meanm = df_chi_meanm.apply(lambda x: '{:02}:{:02}:00'.format(int(x[0]), int(x[1])))
print(df_chi_meanm)
```

## Le nombre total de femmes et d'hommes sur la période.
